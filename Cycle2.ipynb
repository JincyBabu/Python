```
word=input("Enter a Word: ")
vowels=['a','e','i','o','u']
list=[]
for x in word:
    if(x in vowels and x not in list):
      list.append(x)
print("Vowels Present in the given word is: ",list)
```

```
def word_count(txt):
   counts=dict()
   words=txt.split()
   for word in words:
     if word in counts:
       counts[word]+=1
     else:
       counts[word]=1
   return counts
print ( word_count('the quick brown fox jumps over the lazy dog.')) 
```

```
y={'david':45,'alex':20,'christy':35,'jincy':10}
l=list(y.items())
l.sort()
print('Ascending order is:',l)
l=list(y.items())
l.sort(reverse=True)
print('Descending order is:',l)
dict=dict(l)
print("Dictionary",dict)
```
